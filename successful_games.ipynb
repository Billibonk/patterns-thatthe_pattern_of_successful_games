## **Описание данных**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## **2.Подготовка данных**

```python
#Приведем столбцы к нижнему регистру
df.columns = df.columns.str.lower()
```

```python
#Переведем к целочисленному виду столбцы year_of_release и сritic_score
df['year_of_release'] = df['year_of_release'].astype('Int64')
df['critic_score'] = df['critic_score'].astype('Int64')
```

```python
#Обозначим строки с аббревиатурой 'tbd' в столбце с оценкой пользователей
df.query('user_score =="tbd"')
```

Аббревиатура TBD расшифровывается как: 'To Be Determined' - 'Будет определено'. Т.е. это то же самое, что неизвестное значение. Поэтому, можно смело заменить значение 'tbd' на 'NaN'.

```python
#Заменим tbd на NaN
df['user_score'] = df['user_score'].replace('tbd', np.nan, regex=True)
```

```python
#Переведем к целочесленному виду столбец user_score
df['user_score'] = df['user_score'].astype('float64')
```

```python
#Удалим пропущенные строки из столбца name к целочесленному виду столбец user_score
df.dropna(subset=['name'], inplace=True)
df.reset_index(inplace=True,drop=True)
```

```python
#Посмотрим на таблицу с пустыми значениями в столбце year_of_release
df[df['year_of_release'].isna()]
```

```python
#проверим наличие этой игры для других платформ
df.query('name=="LEGO Batman: The Videogame"')
```

```python
# Попробуем востановить пробелы в годах за счет наличия в датасете информации об одних и тех же играх, только выпущенных для разных платформ.
for i in df[df['year_of_release'].isnull() == True].index:  
    df['year_of_release'][i] = df.loc[df['name'] == df['name'][i], 'year_of_release'].max()
```

### Оценка зависимости продаж на примере популярной платформы PS3


```python
#Создадим переменную с актуальным периодом
actual_data = df[df['year_of_release'] >= 2010]
actual_data
```

```python
#Составим график зависимости продаж от отзывов критиков
one_pl = actual_data.query('platform=="PS3"')[['total_sales', 'critic_score', 'user_score']].dropna().reset_index(drop=True)
one_pl.plot(x='critic_score',y='total_sales',kind='scatter',figsize=(15,5), alpha=0.5)
plt.title('Зависимость продаж от отзывов критиков');
plt.xlabel('Оценка критиков')    
plt.ylabel('Объем продаж, млн шт.')
```

```python
#Составим график зависимости продаж от от отзывов пользователей
one_pl.plot(x='user_score',y='total_sales',kind='scatter',figsize=(15,5), alpha=0.5)
plt.title('Зависимость продаж от отзывов пользователей');
plt.xlabel('Оценка критиков')    
plt.ylabel('Объем продаж, млн шт.')
```

```python
one_pl.corr()
```

```python
phik_overview = one_pl.phik_matrix()
phik_overview.round(2)
```

**Вывод:**

Зависимость отзывов критиков на продажи ниже среднего и составляет по коэф. 0.39, по фик ( [Phik (𝜙k) correlation coefficient](https://towardsdatascience.com/phik-k-get-familiar-with-the-latest-correlation-coefficient-9ba0032b37e7)) корреляции составляет 0.30
Зависимость отзывов пользователей на продажи очень низкая и составляет по коэф.  0.07, по фик составляет 0.36
Влияние оценок критиков и пользователей на продажи, не значительное и если смотреть по кореляции фик то зависимость оценок критиков и пользователей практически одинаковая.

## Проверка гипотез

### Гипотеза №1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые 

- Нулевая гипотеза H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Альтернативная гипотеза H₁: Средние пользовательские рейтинги платформ Xbox One и PC различаются

<div class="alert alert-success">
<b>Комментарий ревьюера:</b>

Гипотезы сформулированы верно.
    
    
**Best practies:**
    
<s>Перед проверкой гипотезы рекомендую проверить дисперсии выборок, чтобы корректно определить атрибут `equal_val` в Т–тесте.</s>
    
В том числе стоит визуализировать распределение каждой выборки, чтобы определить нормальное ли у них распределение или нет.
    
Так мы покажем логичность и последовательность наших действий.
    
</div>

```python
# Сохраним в переменных x_one и pc соответствующие данные 
x_one = df.query(' platform == "XOne" ')['user_score'].dropna().to_list()
pc = df.query(' platform == "PC" ')['user_score'].dropna().to_list()
```

```python
np.var(pc) / np.var(x_one)
```

Дисперсии отличаются (около 23,3%), поэтому возьмем equal_var=False

```python
from scipy import stats as st

alpha = 0.05

results = st.ttest_ind(x_one, pc, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза отвергнута - это значит, что и Гипотеза №1 отвергнута, значит средние пользовательские рейтинги платформ Xbox One и PC различны

<div class="alert alert-warning">
<b>Комментарий ревьюера:</b>

В конструкции `if`можно писать условие без скобок. Для логики это неважно, но чем меньше символов в коде — тем легче его читать. Чтобы продемонстрировать коллегам хороший стиль кода, привыкай писать условия без скобок.

</div>

### Гипотеза №2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


- Нулевая гипотеза H₀: Средние пользовательские рейтинги жанров Action и Sports одинаковые.
- Альтернативная гипотеза H₁: Средние пользовательские рейтинги жанров Action и Sports различаются.

```python
# Сохраним в переменных genre_action и genre_sport соответствующие данные 
genre_action = df.query(' genre == "Action" ')['user_score'].dropna().to_list()
genre_sport = df.query(' genre == "Sports" ')['user_score'].dropna().to_list()
```

```python
np.var(genre_action)
```

```python
np.var(genre_sport)
```

```python
np.var(genre_sport) / np.var(genre_action)
```

Дисперсии отличаются (около 27,8%), поэтому возьмем equal_var=False

```python
from scipy import stats as st

alpha = 0.05

results = st.ttest_ind(genre_action, genre_sport, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Не получилось отвергнуть нулевую гипотезу - это значит, что Гипотеза №2 принята, а Средние пользовательские рейтинги жанров Action и Sports одинаковые

Проверили гипотезы по функции ttest_ind, вычисляющей t-тест двух независимых выборок:

- Гипотеза №1: "Средние пользовательские рейтинги платформ Xbox one и PC одинаковые".Отвергаем нулевую гипотезу.
- Гипотеза №2: "Средние пользовательские рейтинги жанров Action и Sports разные". Нулевую гипотезу не удалось опровергнуть.

В Японии совсем другая картина: тут более популярны платформы DS, SNES и 3DS. Следует отметить что платформа PS2 очень входит в топ 3 во всех регионах.
